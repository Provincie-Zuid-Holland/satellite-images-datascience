```python
import glob
import pandas as pd
import geopandas as gpd
import rasterio
from nso_ds_classes.nso_ds_normalize_scaler import scaler_class_all
from annotations.data_preparation import extract_dataframe_pixels_values_from_tif_and_polygons
import os
```

## Voornes Duin

```python
# Set variables
location = "Voornes Duin"
folder = "E:/Data/remote_sensing/satellite-images/"
regex = f"*{location}*ndvi_height_asphalt_crop.tif"
annotation_path = "C:/Users/pzhadmin/Data/remote-sensing/annotations/annotaties_VoornesDuin_2024-01-05_4_labels.geojson"
```

### Prepare data

```python
annotations_polygons_gdf = gpd.read_file(annotation_path)
if annotations_polygons_gdf.crs != "epsg:28992":
        annotations_polygons_gdf = annotations_polygons_gdf.to_crs(epsg=28992)
```

```python
dfs = []
for tif_file in glob.glob(os.path.join(folder, regex)): #path to the satellite image
    tif_file = tif_file.replace("\\","/")
    print(tif_file)
    name_tif_file = tif_file.split("/")[-1].split(".")[0]
    with rasterio.open(tif_file) as dataset:
        dfs += [
            extract_dataframe_pixels_values_from_tif_and_polygons(
                tif_dataset=dataset, polygon_gdf=annotations_polygons_gdf,name_tif_file=name_tif_file
            )
        ]
    
df = pd.concat(dfs)
```

```python
#save dataframe
df.to_parquet("VoornesDuin_polyg2pixel_new.parquet")
```

### Normalise DataFrame through Scalers

```python
df = pd.read_parquet("VoornesDuin_polyg2pixel_new.parquet")
df
```

```python
df_scaled = df.copy()
```

```python
for image_date in df_scaled['date'].unique():
    print(image_date)
    filepath = "../../scalers/{image_date}*{location}*band{{band}}*".format(image_date=image_date, location=location)
    a_normalize_scaler_class_all = scaler_class_all(
        scaler_file_band1 = glob.glob(filepath.format(band=1))[0].replace("\\","/"), \
        scaler_file_band2 = glob.glob(filepath.format(band=2))[0].replace("\\","/"), \
        scaler_file_band3 = glob.glob(filepath.format(band=3))[0].replace("\\","/"), \
        scaler_file_band4 = glob.glob(filepath.format(band=4))[0].replace("\\","/"), \
        scaler_file_band5 = glob.glob(filepath.format(band=5))[0].replace("\\","/"), \
        scaler_file_band6 = glob.glob(filepath.format(band=6))[0].replace("\\","/")
    )
    
    df_scaled[df_scaled['date'] == image_date] = a_normalize_scaler_class_all.transform(
        df_scaled[df_scaled['date'] == image_date], col_names=["r","g","b","i",'ndvi','height']
    )
```

```python
#save scaled dataframe
df_scaled.to_parquet("VoornesDuin_polyg2pixel_scaled_new.parquet")
```
