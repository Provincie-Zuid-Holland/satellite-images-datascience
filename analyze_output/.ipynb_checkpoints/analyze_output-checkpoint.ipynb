```python
import glob
import geopandas as gpd
import pandas as pd
import matplotlib.pylab as plt
import rasterio
from matplotlib import pyplot
import numpy as np
```

```python
date_annotations = [20190601, 20200625, 20200731, 20200915,  20210815,
       20210907, 20220515]
```

```python
gdf = gpd.GeoDataFrame()
```

### Coepelduynen

```python
for date in date_annotations:
  print(date)
  
  file = glob.glob("E:/output/Coepelduynen_segmentations/"+str(date)+"*_v1_3*.shp")[0].replace("\\","/")
  print(file)
  gdf_cur = gpd.read_file(file)
  gdf_cur = gdf_cur.set_crs("EPSG:28992")
  gdf_cur = gdf_cur.to_crs("+proj=cea +units=m")
  gdf_cur['area'] = gdf_cur.area 
  gdf_cur['image'] = file.split("/")[-1]
  gdf = pd.concat([gdf, gdf_cur])

gdf = gdf.drop(['geometry'], axis= 1)
gdf['date'] = gdf.apply( lambda x: x['image'].split("_")[0], axis=1)
```

```python
# English translation
gdf = gdf.replace('Asfalt','Asphalt')
gdf = gdf.replace('Bos','Forest')
gdf = gdf.replace('Gras','Grass')
gdf = gdf.replace('Schaduw','Shadow')
gdf = gdf.replace('Vochtige duinvallei','Wet dune valley')
gdf = gdf.replace('Struweel','Shrub')
gdf = gdf.replace('Zand','Sand')
```

```python
gdf
```

```python
gdf = gdf.join(gdf.groupby(['date']).sum(), on='date', rsuffix='_sum')
gdf['percentage_area'] = gdf.apply(lambda x: (x['area']/x['area_sum'])*100 , axis=1)
gdf.groupby(['date']).sum()
```

```python
gdf
```

```python
gdf[gdf['date'] ==  "20210815"]
```

```python
#colors = {'Asfalt': "grey", 'Bos':"darkgreen", 'Gras':"green", 'Schaduw':"black", 'Struweel':"lightgreen",
#       'Vochtige duinvallei':"lightyellow", 'Zand':"yellow"}


colors ={'Asphalt': "grey", 'Forest':"darkgreen", 'Grass':"lightgreen", 'Shadow':"black", 'Shrub':"mediumseagreen",
       'Wet dune valley':"lightblue", 'Sand':"yellow"}
```

```python
gdf['label'].unique()
```

```python
for alabel in gdf['label'].unique():
    gdf_cur = gdf[gdf['label'] == alabel]
    plt.plot(gdf_cur['date'], gdf_cur['percentage_area'], color=colors[alabel], marker='o')
    plt.title('Area percentage per vegetation type for Coepelduynen', fontsize=14)
    plt.xlabel('Date', fontsize=14)
    plt.ylabel('Area in percentage %', fontsize=14)
    plt.grid(True)

plt.legend(gdf['label'].unique()) 
plt.savefig("time_series_coepelduynen.pdf", format="pdf", bbox_inches="tight")
plt.show()
```

### Voornes duinen

```python
# Replace labels in new save.

for file in glob.glob("E:/output/voornes_duinen/*model2.shp"):
    print(file)
    gdf = gpd.read_file(file)
    gdf = gdf.replace("Grijze duin", "Gras")
    gdf = gdf.replace("Laag gras", "Gras")
    # English translation
    gdf = gdf.replace('Asfalt','Asphalt')
    gdf = gdf.replace('Bos','Forest')   
    gdf = gdf.replace('Gras','Grass')
    gdf = gdf.replace('Schaduw','Shadow')
    gdf = gdf.replace('Vochtige duinvallei','Wet dune valley')
    gdf = gdf.replace('Struweel','Shrub')
    gdf = gdf.replace('Zand','Sand')
    gdf = gdf.dissolve(by='label')
    gdf.to_file(file.replace("\\","/").replace(".shp","_less_labels.shp"))  
```
